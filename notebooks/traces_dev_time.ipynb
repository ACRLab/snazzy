```python
%load_ext autoreload
%autoreload 2
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from pasna_analysis import DataLoader, Experiment

experiment_names = ['20240515', '20240611']

experiments = {}
for exp in experiment_names:
    exp_path = Path.cwd().parent.joinpath('data', exp)
    experiments[exp] = Experiment(DataLoader(exp_path))
```

```python
# to plot metrics by dev_time, I need to get the metric value at a
# sequence of dev_time points. This will, seaborn will group them for me
n = 200
dev_times = [i/n for i in range(int(1.9*n),int(3*n))]
data = {'exp': [], 'dff': [], 'dev_time': []}
for exp in experiments.values():
    for emb in exp.embryos:
        trace = exp.traces[emb.name]
        time_points = emb.get_time_from_DT(dev_times)
        time_points = [int(t/120) for t in time_points]
        for dev_time, t in zip(dev_times, time_points):
            if t <= 0 or t > trace.trim_idx:
                data['dff'].append(0)
            else:
                data['dff'].append(trace.dff[t])
            data['exp'].append(exp.name)
            data['dev_time'].append(dev_time)

sns.set_theme(style='darkgrid')
ax = sns.lineplot(data=data, x='dev_time', y='dff', hue='exp')
```
